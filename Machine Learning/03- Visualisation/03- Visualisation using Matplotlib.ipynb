```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate random numbers
x = np.linspace(0,50,100)
y = x*np.linspace(100,200,100)
```

### Simple Plot

```python
plt.plot(x,y);
```

```python
# SETTING PARAMETERS
plt.plot(x,y);
plt.title('Basic Plot')  # setting the title
plt.xlabel('x-axis')  # label the axis
plt.ylabel('y-axis')
plt.xlim(0,60)  # limit the values
plt.ylim(0,15000)
```

```python
plt.plot(x,y,c='red',linestyle='--',linewidth=5);
```

```python
plt.plot(x,y,c='r',markersize=9,marker='*');
```

```python
plt.plot(x,y,c='g',markersize=12,marker='1',label='normal');
plt.grid(True)
plt.legend();
```

### Multiple Plots in one figure

```python
x= np.arange(0,1500,100)
plt.plot(np.sin(x),label='sin function x')
plt.plot(np.cos(x),label='cos functon x')
plt.legend(loc='upper right');
```

### Sub Plots

```python
x = np.random.rand(50)
y = np.sin(x*2)
```

```python
'''
create an empty subplot returning 
a tuple of figure and set of subplots
'''
fig,axs = plt.subplots()    
axs.plot(y);
```

```python
# create 'n' subplots
fig , axs = plt.subplots(nrows=2)

# assign data to plots
x= np.linspace(0,100,10)
axs[0].plot(x,np.sin(x**2))
axs[1].plot(x,np.cos(x**2))

# add a title to the subplot
fig.suptitle('Vertical Sub-Plots')
```

```python
# create 'n' subplots
fig , axs = plt.subplots(ncols=2)

# assign data to plots
x= np.linspace(0,100,10)
axs[0].plot(x,np.sin(x**2))
axs[1].plot(x,np.cos(x**2))

# add a title to the subplot
fig.suptitle('Horizontal Sub-Plots')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

# add the data referring to row and column
axs[0,0].plot(x, x**2,'g')
axs[0,1].plot(x, x**3,'r')
axs[1,0].plot(x, np.sin(x**2),'b')
axs[1,1].plot(x, np.cos(x**2),'k')

# add title
fig.suptitle('Matrix Sub-Plots')
```

### Figure Object

```python
width = 4   # in inches
height = 3
# Figure object with figure size
fig = plt.figure(figsize=(width,height),facecolor='y',edgecolor='r',linewidth=3) 

# create data
x = np.arange(1,11)
y = x**2

plt.plot(x,y);
```

### Axes Object

```python
y = np.arange(1,31,5)
x1 = [1,10,20,30,45,55]
x2 = [1,32,45,80,90,122]

# create figure
fig = plt.figure()

# add axes - dimensions [left, bottom, width, height]
ax = fig.add_axes([0,0,2,1])
l1 = ax.plot(x1,y,'ys-')
l2 = ax.plot(x2,y,'go--')

ax.legend(labels=('line 1','line 2'),loc = 'lower right')
ax.set_title('Usage of Axes')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
plt.show()
```

### Bar Graph

```python
# create data
subjects = ['maths','english','science','social','computer']
marks = [70,80,50,30,78]

# create bar
plt.bar(x=subjects,height=marks,color='g',width=0.5,bottom=10,align='center',tick_label=subjects,linewidth=2,yerr=np.std(marks))

# display
plt.show()
```

```python
plt.barh(subjects,marks,xerr=np.std(marks));
```

### Pie Chart

```python
# create data
tickets_closed = [10,20,8,35,30,25]
agents = ['A','B','C','D','E','F']

# move out wedge or widget
explode = [0.2,0.1,0,0,0,.3]
# create pie chart
# %.1f%% -- passing a string to percetage upto precision of 1 and display % sign
plt.pie(x=tickets_closed,labels=agents,explode=explode,autopct='%1.1f%%');
```

### Scatter Plot

```python
# create data
x = np.random.randn(1000)
y = np.random.randn(1000)

# plotting
plt.scatter(x,y)
plt.show()


# additional parameters
size =100*np.random.randn(1000)
colors = 100*np.random.randn(1000)

plt.scatter(x,y,s=size,c=colors,marker='*',alpha=0.7)
plt.show()
```

### Saving the plots

```python
items = [5,10,20,25,30,40]
x = np.arange(6)
y = np.arange(6)
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(x,y,label=items)
plt.title('Saving as Image')
ax.legend()
fig.savefig('saveimage.png')
```

```python
# DISPLAYING SAVED IMAGES
import matplotlib.image as mpimg
image = mpimg.imread('saveimage.png')
plt.imshow(image);
```
