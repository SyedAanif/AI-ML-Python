```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(10,4),
                  index=pd.date_range('10/28/2020',periods=10),
                  columns=list('ABCD'))
df
```

### Line Plot

```python
df.A.plot();
```

### Bar Plot

```python
df.A.plot.bar(); # vertical bar plot
```

```python
df.B.plot.bar(stacked=True);
```

```python
# horizontal bar plot
df.C.plot.barh();
```

### Histogram

```python
df.D.plot.hist(bins=20);
```

```python
df.A.hist();
```

```python
df.hist();
```

### Scatter Plot

```python
df.plot.scatter(x='A',y='B');
```

### Pie Chart

```python
df[df>0].plot.pie(y='C');
```
